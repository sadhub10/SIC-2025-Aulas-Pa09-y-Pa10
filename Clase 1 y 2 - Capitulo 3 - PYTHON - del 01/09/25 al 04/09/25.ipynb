As铆, el closure es ese "paquete de memoria" que la funci贸n interna mantiene de la funci贸n externa, permiti茅ndole operar con datos que ya no est谩n en su 谩mbito inmediato.

### 1. Funciones: "La M谩quina de Hacer Jugo" 

```python
# Definici贸n de la funci贸n 'hacer_jugo'
# 'fruta' y 'azucar' son los par谩metros (los ingredientes de nuestra receta)
def hacer_jugo(fruta, azucar_cucharadas):
    # El cuerpo de la funci贸n: los pasos para hacer el jugo
    print(f"Mezclando {fruta} con {azucar_cucharadas} cucharadas de az煤car.")

    # La sentencia 'return': el producto final
    return f"Aqu铆 est谩 tu jugo de {fruta}."

# Llamada a la funci贸n
# "naranja" y 2 son los argumentos (los valores concretos que pasamos)
jugo_listo = hacer_jugo("naranja", 2)

# Usamos el valor que la funci贸n nos devolvi贸
print(jugo_listo)
```

### 2. Closures: "El Fot贸grafo que Recuerda tu Nombre" 

```python
def contratar_fotografo(nombre_cliente):
    """
    Funci贸n exterior (el fot贸grafo) que recibe el nombre del cliente.
    'nombre_cliente' es la variable del entorno que el closure "recuerda".
    """

    def tomar_foto(evento):
        """
        Funci贸n anidada (el closure) que toma la foto.
        Tiene acceso a 'nombre_cliente' incluso fuera de su funci贸n original.
        """
        return f"Foto de {evento} tomada para {nombre_cliente}."

    # La funci贸n exterior retorna la funci贸n anidada (el closure)
    return tomar_foto

# 'fotografo_para_laura' es ahora nuestro closure
fotografo_para_laura = contratar_fotografo("Laura")

# Llamamos al closure, que recuerda el nombre "Laura"
print(fotografo_para_laura("la boda"))
# Salida: Foto de la boda tomada para Laura.
```

# **POO**

**La POO se basa en dos conceptos principales: clases y objetos.**

## OTROS MTODOS DOUNDER COMUNES

### __repr__ (Representaci贸n del Desarrollador)
Similar a __str__, pero su objetivo es dar una representaci贸n inequ铆voca del objeto que, idealmente, se podr铆a usar para recrear la instancia. Se usa para la depuraci贸n y es la representaci贸n por defecto si __str__ no est谩 definido.

```python
class Punto:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __repr__(self):
        return f"Punto(x={self.x}, y={self.y})"

p = Punto(6, 4)
print(repr(p))  # Salida: Punto(x=3, y=4)
```

### __len__ (Longitud)
Este m茅todo permite que tu objeto sea compatible con la funci贸n len(). Es ideal para clases que son colecciones de datos, como una lista o un diccionario.

```python
class Pila:
    def __init__(self):
        self.items = []

    def __len__(self):
        return len(self.items)

    def _push(self, item):
        self.items.append(item)

pila = Pila()
pila._push(10)
pila._push(20)
pila._push(30)
pila._push(60)
print(len(pila))  # Salida: 2
```

### __eq__ (Comparaci贸n de Igualdad)
Este m茅todo te permite definir c贸mo se comparan dos objetos de tu clase con el operador ==.

Ejemplo de implementaci贸n:

```python
class Coche:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo

    def __eq__(self, otro):
        # Dos coches son iguales si su marca y modelo son los mismos
        return self.marca == otro.marca and self.modelo == otro.modelo

coche1 = Coche("Bmw", "Model 5")
coche2 = Coche("Bmw", "Model 5")
coche3 = Coche("Rold Royce", "Model 1")

print(coche1 == coche2)
print(coche1 == coche3)
```

### __add__ (Suma)
Este m茅todo te permite definir c贸mo se comporta el operador + con tu clase, lo que puede ser 煤til para combinar objetos.

Ejemplo de implementaci贸n:

```python
class Vector:
    def __init__(self, x, y , L):
        self.x = x
        self.y = y
        self.L = L

    def __add__(self, otro):
        # Suma las componentes de dos vectores
        return Vector(self.x + otro.x , self.y + otro.y , self.L + otro.L)

    def __str__(self):
        return f"({self.x}, {self.y}, {self.L})"

v1 = Vector(2, 3, 1)
v2 = Vector(6, 5, 2)
v3 = v1 + v2
v4 = v1 + v2 + v3
v5 = v1 + v2 + v3 + v4
print(v3)
print(v5)
```
