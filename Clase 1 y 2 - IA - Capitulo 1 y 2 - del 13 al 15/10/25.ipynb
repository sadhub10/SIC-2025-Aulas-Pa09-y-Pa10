## ¿Qué es un Vector?

Un vector es un objeto que tiene magnitud (longitud) y dirección. En Data Science, un vector a menudo representa una observación de datos (ej: un cliente con variables como edad, ingresos, etc.).

Vectores (Numpy)
Usaremos la biblioteca NumPy para manejar vectores y matrices.

```python
import numpy as np

# 1. Definición de dos vectores de ejemplo (Punto A y Punto B)
# Vector A: edad=30, ingresos=50000, Vector B: edad=40, ingresos=60000
A = np.array([30, 50000])
B = np.array([40, 60000])

# 2. Norma Euclidiana (Magnitud o Distancia desde el origen)
# ||A|| = sqrt(30^2 + 50000^2)
norma_A = np.linalg.norm(A)
print(f"Norma Euclidiana de A: {norma_A:.2f}")

# 3. Producto Interno (Producto Punto)
# Se usa para calcular la Similitud
producto_punto = np.dot(A, B)
print(f"Producto Punto de A y B: {producto_punto}")

# 4. Similitud del Coseno (Aplicación del Producto Punto)
# Cos(theta) = (A . B) / (||A|| * ||B||)
norma_B = np.linalg.norm(B)
similitud_coseno = producto_punto / (norma_A * norma_B)
print(f"Similitud del Coseno entre A y B: {similitud_coseno:.4f} (Muy cercanos a 1, son similares)")
```

### Funcionamiento lógico:

1. Importación y Definición de Vectores

* import numpy as np: Importa la librería NumPy, que es la base para la computación numérica eficiente en Python. Se le da el alias np.

* Vectores A y B: Se definen dos vectores (arreglos de NumPy) que representan dos puntos de datos en un espacio bidimensional.

* A y B son ejemplos de perfiles de clientes donde la primera coordenada es la edad y la segunda son los ingresos.

2. Norma Euclidiana (Magnitud)

* Concepto: La Norma Euclidiana (∥A∥), también conocida como magnitud o norma L2, representa la distancia del vector desde el origen (0,0) del sistema de coordenadas.

* Función: np.linalg.norm(A) calcula esta magnitud usando el Teorema de Pitágoras generalizado:  

* Propósito: Se usa para normalizar vectores, que es un paso crucial en el cálculo de la Similitud del Coseno.

3. Producto Interno (Producto Punto)

* Concepto: El Producto Interno (o Producto Punto) es una operación que toma dos vectores y devuelve un único número escalar. Se calcula multiplicando los elementos correspondientes de los dos vectores y sumando los resultados.

* Función: np.dot(A, B) realiza este cálculo.

* Propósito: Este valor es la base para calcular la Similitud del Coseno.

4. Similitud del Coseno (Similitud)

* Concepto: La Similitud del Coseno mide el ángulo entre dos vectores, independientemente de su magnitud. Es una medida popular de similitud en minería de textos y sistemas de recomendación.

 Interpretación del Resultado:

* Un valor de 1 indica que los vectores apuntan exactamente en la misma dirección (son idénticos en orientación, o muy similares).

* Un valor de 0 indica que los vectores son ortogonales (sin relación).

* Un valor de -1 indica que los vectores son opuestos.

El resultado cercano a 1 en tu ejemplo muestra que los perfiles de datos A y B (30 años/50k ingresos y 40 años/60k ingresos) tienen una orientación muy similar, por lo que se consideran muy parecidos en este contexto.

## ¿Qué es una Matriz

Una matriz es unarreglo rectangular de números, variables o símbolos, distribuidos en filas (horizontalmente) y columnas (verticalmente), que se utilizan en matemáticas y otras disciplinas para organizar datos y describir sistemas.

```python
import numpy as np

# Matriz A (ejemplo: 2 observaciones, 3 características)
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

# Matriz de Pesos W (ejemplo: 3 características, 2 neuronas de salida)
W = np.array([
    [0.1, 0.4],
    [0.2, 0.5],
    [0.3, 0.6]
])

# 1. Transpuesta de A (Filas se vuelven Columnas)
A_transpuesta = A.T
print("Matriz A Transpuesta:\n", A_transpuesta)

# 2. Producto Matricial (La operación clave: (2x3) * (3x2) = (2x2))
# El operador '@' o np.dot() realiza el producto matricial
producto_matricial = A @ W
print("\nResultado del Producto Matricial (A @ W):\n", producto_matricial)
```

## ¿Qué es una Derivada?

Una derivada es latasa de cambio instantáneo de una función en un punto específico, que se interpreta como la pendiente de la recta tangente a la gráfica de la función en ese punto.

```python
from sympy import symbols, diff, exp

# 1. Definir el símbolo de la variable (x)
x = symbols('x')

# 2. Definir una función simple: f(x) = x^2 + 2x + 1
f_x = x**2 + 2*x + 1

# 3. Calcular la Derivada de f(x): f'(x) = 2x + 2
derivada_f_x = diff(f_x, x)
print(f"La derivada de f(x) = {f_x} es f'(x) = {derivada_f_x}")

# 4. Derivada de la Función Sigmoide (Aplicación en Deep Learning)
# La Función Sigmoide: sigma(x) = 1 / (1 + e^(-x))
# Su derivada es: sigma'(x) = sigma(x) * (1 - sigma(x))
sigma_x = 1 / (1 + exp(-x))

# Calcular la derivada de la función sigmoide
derivada_sigma_x = diff(sigma_x, x)
print(f"\nLa derivada de la Función Sigmoide es:\nσ'(x) = {derivada_sigma_x}")
```
