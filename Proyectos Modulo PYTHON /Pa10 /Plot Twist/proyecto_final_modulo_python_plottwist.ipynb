# Diccionario de Terminologas Mdicas

| **Trmino**                                | **Definicin**                                                                                                                                                                                                                                       |
| ------------------------------------------ | ---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **Angina**                                 | Dolor o molestia en el pecho causada por una disminucin del flujo sanguneo al corazn. Puede ser *tpica* (con patrn predecible) o *atpica* (sntomas menos claros).                                                                             |
| **Angina inducida por ejercicio (Exang)**  | Episodio de dolor torcico que aparece al realizar esfuerzo fsico, debido a una insuficiencia en el flujo sanguneo cardaco durante la actividad.                                                                                                  |
| **Colesterol (Chol)**                      | Sustancia grasa que circula por la sangre. Niveles altos pueden provocar acumulacin de placas en las arterias y aumentar el riesgo de enfermedad cardaca.                                                                                          |
| **Depresin del segmento ST (Oldpeak)**    | Alteracin en el electrocardiograma que indica posible isquemia cardaca (falta de oxgeno en el corazn).                                                                                                                                           |
| **Electrocardiograma en reposo (RestECG)** | Prueba que registra la actividad elctrica del corazn mientras el paciente est en reposo. Permite detectar arritmias o lesiones cardacas.                                                                                                         |
| **Fluoroscopa (CA)**                      | Tcnica de rayos X en tiempo real que permite observar el funcionamiento del corazn y vasos sanguneos; en este caso, mide cuntos vasos principales presentan flujo anormal.                                                                       |
| **Glucemia en ayunas (FBS)**               | Nivel de azcar en la sangre tras un periodo sin ingerir alimentos. Valores elevados pueden indicar riesgo de diabetes o sndrome metablico.                                                                                                        |

| **Presin arterial en reposo (Trestbps)**  | Fuerza que ejerce la sangre contra las paredes de las arterias cuando el corazn est en reposo entre latidos.                                                                                                                                       |
| **Segmento ST**                            | Parte del electrocardiograma que representa el periodo entre la despolarizacin y repolarizacin del ventrculo; cambios en su forma indican dao o estrs cardaco.                                                                                 |
| **Talio / Thalassemia (Thal)**             | En este contexto se refiere a un *test de talio*, una prueba de esfuerzo con un istopo radiactivo que evala el flujo sanguneo al corazn. Los resultados pueden indicar defectos fijos (dao permanente) o reversibles (falta temporal de flujo). |
| **Pendiente del ST (Slope)**               | Describe la inclinacin del segmento ST durante el esfuerzo. Su forma puede indicar si el corazn recibe suficiente oxgeno durante el ejercicio.                                                                                                    |
| **Frecuencia cardaca mxima (Thalach)**   | El nmero ms alto de latidos por minuto que una persona alcanza durante una prueba de esfuerzo. Es un indicador de la capacidad cardiovascular.                                                                                                     |
| **Vasos principales (CA)**                 | Arterias coronarias ms grandes que suministran sangre al corazn; su bloqueo o estrechamiento puede causar angina o infarto.                                                                                                                        |

## La hipertensin se asocia con la presencia de enfermedad cardaca?



```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import chi2_contingency

# 1Crear variable categrica 'hipertenso'
df['hipertenso'] = (df['trestbps'] > 130).astype(int)  # 1 = hipertenso, 0 = no hipertenso

# Ver distribucin general
print(df['hipertenso'].value_counts())

# Tabla de contingencia
tabla = pd.crosstab(df['hipertenso'], df['target'])
print("\nTabla de contingencia:")
print(tabla)

# Prueba Chi-cuadrado
chi2, p, dof, expected = chi2_contingency(tabla)
print(f"\nChi-cuadrado = {chi2:.2f}")
print(f"p-valor = {p:.4f}")

if p < 0.05:
    print(" Existe una asociacin significativa entre hipertensin y enfermedad cardaca.")
else:
    print(" No se encontr una asociacin estadsticamente significativa entre hipertensin y enfermedad cardaca.")

# Grfico comparativo
sns.countplot(data=df, x='hipertenso', hue='target', palette=['#8FD694', '#E57373'])
plt.title('Relacin entre hipertensin y enfermedad cardaca')
plt.xlabel('Hipertenso (1 = S, 0 = No)')
plt.ylabel('Cantidad de pacientes')
plt.legend(title='Enfermedad', labels=['No','S'])
plt.show()

```

Se encontr una asociacin significativa (p = 0.0058) entre hipertensin y enfermedad cardaca, lo que sugiere que la presin arterial elevada podra influir en el riesgo cardiovascular dentro de esta muestra.

# Correlacin entre variables.



### **Anlisis de Correlacin entre Variables**

Para obtener una visin general de las relaciones lineales entre las distintas variables numricas del conjunto de datos, generamos una **matriz de correlacin**. La matriz se visualiza con un  **mapa de calor (heatmap)**, una herramienta grfica que utiliza una escala de colores para representar la fuerza y la direccin de las correlaciones.

En este mapa de calor:
*   Los colores ms **claros o clidos** indican una **correlacin positiva fuerte** (cuando una variable aumenta la otra tambien).
*   Los colores ms **oscuros o fros** indican una **correlacin negativa fuerte** (cuando el valor una variable aumenta la otra debe disminuir).
*   Los colores **intermedios** sugieren una **correlacin dbil o nula**.

Nuestro principal inters se centra en la ltima fila/columna, que muestra la correlacin de cada variable con nuestra variable objetivo (`target`), dndonos pistas iniciales sobre los factores ms predictivos. Adicionalmente, este anlisis nos permite detectar posible **multicolinealidad** (alta correlacin entre variables predictoras).


```python

plt.figure(figsize=(12, 10))
sns.heatmap(df.corr())
plt.title('Mapa de Calor de Correlacin entre Variables', fontsize=16)
plt.show()
```


### **Anlisis del Grfico de Correlacin**

El mapa de calor nos permitio identificar:

## 1. Lgica del Sistema de Recomendaciones

```python
import gradio as gr
import pandas as pd

# Mapeo de etiquetas legibles a los valores numricos del dataset
SEX_MAP = {"Mujer (0)": 0, "Hombre (1)": 1}
CP_MAP = {"0 - Angina tpica": 0, "1 - Angina atpica": 1, "2 - Dolor no anginoso": 2, "3 - Asintomtico": 3}
FBS_MAP = {"0 - Normal (120 mg/dl)": 0, "1 - Alta (>120 mg/dl)": 1}
RESTECG_MAP = {"0 - Normal": 0, "1 - Anomala ST-T": 1, "2 - HVI probable": 2}
EXANG_MAP = {"0 - No": 0, "1 - S": 1}
SLOPE_MAP = {"0 - Descendente": 0, "1 - Plano": 1, "2 - Ascendente": 2}
THAL_MAP = {"0 - No aplica": 0, "1 - Defecto fijo": 1, "2 - Defecto reversible": 2, "3 - Normal": 3}

def recomendar_salud_cardiaca(age, sex_lbl, cp_lbl, trestbps, chol, fbs_lbl, thalach, exang_lbl, oldpeak, slope_lbl, thal_lbl):
    """
    Calcula un puntaje de riesgo y genera recomendaciones basadas en guas mdicas generales
    y los hallazgos del anlisis del dataset.
    """
    risk_score = 0
    recommendations = []

    # Mapear entradas de texto a valores numricos
    sex = SEX_MAP[sex_lbl]
    cp = CP_MAP[cp_lbl]
    fbs = FBS_MAP[fbs_lbl]
    exang = EXANG_MAP[exang_lbl]
    slope = SLOPE_MAP[slope_lbl]
    thal = THAL_MAP[thal_lbl]

    # --- Lgica de Puntuacin de Riesgo (1 punto por cada factor de riesgo) ---

    # 1. Edad y Sexo
    if (sex == 1 and age > 45) or (sex == 0 and age > 55):
        risk_score += 1
        recommendations.append(" **Edad/Sexo:** La edad es un factor de riesgo natural. Es importante mantener controles mdicos regulares, especialmente a partir de los 45 aos para hombres y 55 para mujeres.")

    # 2. Presin Arterial (trestbps)
    if trestbps > 130:
        risk_score += 1
        recommendations.append(f" **Presin Arterial ({trestbps} mm Hg):** Tu presin arterial en reposo est por encima del nivel ptimo (130 mm Hg). Considera reducir el consumo de sal, hacer ejercicio regularmente y consultar a un mdico.")

    # 3. Colesterol (chol)
    if chol > 240:
        risk_score += 1
        recommendations.append(f" **Colesterol ({chol} mg/dl):** Tu nivel de colesterol es alto (superior a 240 mg/dl). Una dieta baja in grasas saturadas y rica en fibra puede ayudar a controlarlo.")

    # 4. Angina Inducida por Ejercicio (exang)
    if exang == 1:
        risk_score += 1
        recommendations.append(" **Angina por Ejercicio:** Experimentar dolor de pecho durante el ejercicio es una seal de alerta importante. Es fundamental que consultes a un cardilogo.")

    # 5. Depresin del ST (oldpeak)
    if oldpeak > 1.0:
        risk_score += 1
        recommendations.append(f" **Depresin del ST ({oldpeak}):** Un valor de 'oldpeak' elevado sugiere una respuesta anormal del corazn al ejercicio. Este es un indicador que tu mdico debe evaluar en detalle.")

    # 6. Frecuencia Cardaca Mxima (thalach)
    predicted_max_hr = 220 - age
    if thalach < (predicted_max_hr * 0.85): # Si es menor al 85% de la esperada
        risk_score += 1
        recommendations.append(f" **Frecuencia Cardaca Mxima ({thalach} lpm):** Tu frecuencia cardaca mxima es ms baja de lo esperado para tu edad. Esto puede indicar una menor capacidad cardiovascular. El ejercicio aerbico regular puede ayudar a mejorarla.")

    # 7. Test de Talio (thal)
    if thal == 2: # Defecto reversible
        risk_score += 2 # Se le da ms peso por ser un predictor fuerte
        recommendations.append(" **Test de Talio (Defecto Reversible):** Este resultado es un indicador muy fuerte de problemas de flujo sanguneo al corazn. Es crucial que un especialista evale este hallazgo.")
    elif thal == 1: # Defecto fijo
        risk_score += 1
        recommendations.append(" **Test de Talio (Defecto Fijo):** Este resultado puede indicar una cicatriz de un evento cardaco previo. Es importante un seguimiento mdico continuo.")


    # --- Determinacin del Nivel de Riesgo ---
    if risk_score <= 2:
        risk_level = "Riesgo Bajo"
        if not recommendations:
            recommendations.append("Felicidades! Tus indicadores se encuentran dentro de rangos saludables. Contina con un estilo de vida activo y una dieta balanceada.")
    elif risk_score <= 4:
        risk_level = "Riesgo Moderado"
        recommendations.insert(0, "Se han identificado algunos factores de riesgo. Es un buen momento para adoptar hbitos ms saludables y considerar una visita preventiva al mdico.")
    else:
        risk_level = "Riesgo Alto"
        recommendations.insert(0, "Se han identificado mltiples factores de riesgo importantes. Te recomendamos encarecidamente que consultes a un mdico o cardilogo para una evaluacin completa.")

    return risk_level, "\n\n".join(recommendations)
```

##2. Interfaz Grfica con Gradio

```python
# Creacin de la interfaz de Gradio
with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown("# Predictor de Riesgo Cardaco y Recomendaciones")
    gr.Markdown(
        "Introduce tus datos para obtener una estimacin de tu riesgo cardiovascular y recibir recomendaciones personalizadas. "
        "**Aviso:** Esta es una herramienta educativa y no reemplaza el diagnstico mdico profesional."
    )

    with gr.Row():
        with gr.Column():
            gr.Markdown("### Datos Fisiolgicos")
            age_input = gr.Slider(29, 80, value=54, label="Edad (aos)")
            trestbps_input = gr.Slider(90, 200, value=130, label="Presin Arterial en Reposo (mm Hg)")
            chol_input = gr.Slider(120, 600, value=240, label="Colesterol (mg/dl)")
            thalach_input = gr.Slider(70, 210, value=150, label="Frecuencia Cardaca Mxima")
            oldpeak_input = gr.Slider(0.0, 6.5, value=1.0, step=0.1, label="Depresin del ST (oldpeak)")

        with gr.Column():
            gr.Markdown("### Historial y Pruebas")
            sex_input = gr.Dropdown(list(SEX_MAP.keys()), value="Hombre (1)", label="Sexo")
            cp_input = gr.Dropdown(list(CP_MAP.keys()), value="2 - Dolor no anginoso", label="Tipo de Dolor de Pecho")
            fbs_input = gr.Dropdown(list(FBS_MAP.keys()), value="0 - Normal (120 mg/dl)", label="Glucemia en Ayunas")
            exang_input = gr.Dropdown(list(EXANG_MAP.keys()), value="0 - No", label="Angina Inducida por Ejercicio")
            slope_input = gr.Dropdown(list(SLOPE_MAP.keys()), value="1 - Plano", label="Pendiente del Segmento ST")
            thal_input = gr.Dropdown(list(THAL_MAP.keys()), value="2 - Defecto reversible", label="Resultado del Test de Talio")
            restecg_input = gr.Dropdown(list(RESTECG_MAP.keys()), visible=False) # Variable no usada en la lgica, pero necesaria para la funcin

    submit_btn = gr.Button("Analizar Riesgo", variant="primary")

    with gr.Row():
        risk_output = gr.Textbox(label="Nivel de Riesgo Estimado")
        recommendations_output = gr.Textbox(label="Recomendaciones Personalizadas", lines=10)

    submit_btn.click(
        fn=recomendar_salud_cardiaca,
        inputs=[
            age_input, sex_input, cp_input, trestbps_input, chol_input, fbs_input,
            thalach_input, exang_input, oldpeak_input, slope_input, thal_input
        ],
        outputs=[risk_output, recommendations_output]
    )

demo.launch(debug=True)
```
